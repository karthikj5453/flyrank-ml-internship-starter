## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# This cell is for CODE (numbers, a query, a check).
## The Question

### Research Question

Which webpages should be reviewed first for content refresh based on historical search and engagement signals?

### Decision

Decide which pages deserve review before their performance declines further.

### Action

SEO teams can refresh content, improve metadata, expand articles, update outdated information, or monitor pages with high predicted risk.

### Unit of Analysis

One webpage (one row in the dataset).

### Output

A ranked refresh score indicating the priority for reviewing each webpage.

### Cost of a Wrong Recommendation

False Positive:
A team spends time refreshing a page that was unlikely to decline.

False Negative:
An important page that is actually declining is ignored, resulting in lost traffic and missed business opportunities.

### Why Machine Learning?

Large websites may contain thousands of webpages, making manual review difficult. Machine learning can identify patterns in historical search and engagement metrics to prioritize the pages most likely to benefit from review.
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
```
